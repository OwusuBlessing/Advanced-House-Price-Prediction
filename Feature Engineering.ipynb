### Missing values

```python
# get categorical features wiht nan values

features_nan_categorical = [feature for feature in df.columns if df[feature].isnull().sum() > 1 and df[feature].dtypes 
                            == "object"]
for feature in features_nan_categorical:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))

```

```python
# replace missing value in categorical feature
def replace_nan_cat(df,features_nan_categorical):
        data = df.copy()
        df[features_nan_categorical] = df[features_nan_categorical].fillna("Missing")
        return df
df = replace_nan_cat(df,features_nan_categorical)  
df.head()
        
```

```python
# check for numerical variables that contain missing values
numerical_nan = [feature for feature in df.columns if df[feature].isnull().sum() > 1 and df[feature].dtypes 
                            != "object"]
for feature in numerical_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))

```

```python
# replacing numerical variables with nan
for feature in numerical_nan:
    # replace by using median value\
    median_value = df[feature].median()
    # create new features to indicate nan values 1 for nan and 0 ortherwise
    df[feature+"nan"]= np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace = True)
df[numerical_nan].isnull().sum()
  
```

### Temporal variable

```python
# Day time variables are temporal variables
# covert temporable into more useful form
for feature in ["YearBuilt","YearRemodAdd","GarageYrBlt"]:
    df[feature]=df["YrSold"]-df[feature]

    
```

## Handling Rare Categorical features

```python
# remove categorical features which present less than 1% of data observation
categorical_features = [feature for feature in df.columns if df[feature].dtype == "object" and feature!="SalePrice"]
categorical_features






```

```python
# get % of each categorical data
for feature in categorical_features:
    #if feature != "SalePrice":
        temp = df.groupby(feature)["SalePrice"].count()/len(df)
        temp_df = temp[temp > 0.01].index
        df[feature] = np.where(df[feature].isin(temp_df),df[feature],"Rare_var")

    
df.head()
        
```
